座標位移 ── 資訊中心化

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 以下有五組數據點 : x 為 x座標，y為 y座標
x = np.array([0.2, 3.4, 5.4, 2.8, 7.7])
y = np.array([2.4, 4.6, 6.8, 8.1, 10.2])
```

```python
plt.scatter(np.mean(x), np.mean(y))
plt.scatter(x[3],y[3])
plt.text(x[3]+0.3, y[3], '(%.2f,%.2f)'%(x[3],y[3]), fontsize=15)
plt.text(np.mean(x)+0.3, np.mean(y), 'center = (%.2f,%.2f)'%(np.mean(x), np.mean(y)), fontsize=15)
plt.plot(x,y)
plt.title('non_mid')
plt.show()
```

```python
x = x - np.mean(x)
y = y - np.mean(y)
plt.scatter(0,0)
plt.scatter(x[3],y[3])
plt.text(x[3]+0.3, y[3], '(%.2f,%.2f)'%(x[3],y[3]), fontsize=15)
plt.text(0.3, 0, 'center = (0,0)', fontsize=15)
plt.plot(x,y)
plt.title('mid')
plt.show()
```
