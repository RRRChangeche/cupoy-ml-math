### 檢驗最大最小值

首先考慮 $$f:[-3,3] \rightarrow R，$$且$$f(x)=x^3-3x$$最小值或最大值發生在哪裡?

$$f'(x)=3x^2-3=0\\
\Rightarrow x = \pm 1$$

```python
def f(x):
    return x**3-3*x
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 2)

x = np.linspace(-3,3)
ax[0].plot(x[25:],f(x[25:]),label='x^3-3x')
ax[0].scatter(1,f(1),s=50)
ax[0].scatter(3,f(3),s=50)
ax[0].text(0,f(0),'minimum value is %d'%f(1),fontsize=11)
ax[0].text(0,f(2.9),'maximum value is %d'%f(3),fontsize=11)
ax[0].legend()


ax[1].plot(x[:25],f(x[:25]),label='x^3-3x')
ax[1].scatter(-1,f(-1),s=50)
ax[1].scatter(-3,f(-3),s=50)
ax[1].text(-2,f(0),'maximum value is %d'%f(-1),fontsize=11)
ax[1].text(-2.8,f(-2.9),'maximum value is %d'%f(-3),fontsize=11)
ax[1].legend()

fig.tight_layout()
plt.show()
```

```python
x = np.linspace(-10,10)  # 可調整大小，會發現極值與範圍呈正比(愈大或愈小)。
plt.plot(x,f(x),label='x^3-3x')
plt.legend()
plt.show()
```

結論 : 
    我們可以說在 0 到 3 之間([-3,3])，f 的最大值發生在 x=3，最大值為 18; f 的最小值發生在 x=1，最小值為 -2。
    同樣地，在 -3 到 0 之間([-3,3])，f 的最大值發生在 x=-1，最大值為 2; f 的最小值發生在 x=-3，最小值為 -18。
    但是，在全域中，沒有極值(極值不存在)。
    
    所以，討論極值的定義域很重要!
