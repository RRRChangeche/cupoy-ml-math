作業2：找出矩陣找出矩陣對角元素，建立對角矩陣

```python
import numpy as np
np.random.seed(3)
x = np.array([[1, 2, 3, 4],
        [4, 3, 2, 1],
        [4, 3, 2, 1],
        [2, 2, 1, 2]], dtype = np.int16)
z = np.diag(x)
print ("對角線元素：", z)
y = np.diag(z)
print ("還原為對稱(角)矩陣:\n",y)
```

作業3：產生與確認正交(Orthogonal)矩陣

```python
from scipy.stats import ortho_group # 需要 scipy 版本至少 0.18
import numpy as np
#產生一3x3正交矩陣
m = ortho_group.rvs(dim=4, random_state=42)
print(m)
```

```python
#確認正交姓
np.set_printoptions(suppress=True) #把太小的浮點數印成0
l1 = np.matmul(m, m.T)
print ("matmul(m, m.T)")
print(l1)
print()
print ("matmul(m.T, m)")
l2 = np.matmul(m.T, m)
print(l2)
```
