作業目的 : 藉由低維度的資料熟悉梯度下降法，注意，gradient在單變數中可視為一階微分(方向為左或右)。
    
作業題目 : 利用梯度下降法與下面給出的函式，找出以下函數的最小值，並觀察不同學習率(alpha)對於收斂性的影響。(提示 : 不斷更新 x ，使得 f(x) 愈來愈小)

假設一函數 $$f(x)=x^2$$

```python
def f(x):
    return x**2
```

```python
def diff(fcn):
    h=1e-5
    return lambda x : (fcn(x+h)-fcn(x))/h
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5)
plt.plot(x,f(x))
plt.show()
```

### 利用梯度下降法，找出下列函數的最小值

提示 : 在曲線上隨便給一個點

```python
plt.plot(x,f(x))
plt.scatter(x[10],f(x[10]), s=50,color='r')
plt.show()
```

```python
w = x[10]
# alpha = 0.002
# alpha = 0.005
alpha = 0.02

for i in range(200):
    w = w - alpha*diff(f)(w)

    if i%50 == 0:
        print(f"The {i} iteration")
        plt.plot(x,f(x))
        plt.scatter(w,f(w), s=50,color='r')
        plt.show()
    else:
        pass
```
