# **# Numpy NDArray 基本運算**

```python
A = np.array([[3,2],[1,0]])
B = np.array([[3,1],[2,4]])
#A B 相加
A + B
```

```python
#A B 元素乘法
A * B
```

```python
#A B 乘法
np.matmul(A, B)
A @ B
```

```python
#A B 點乘 (elementwise multiplication)
A.dot(B)
```

```python
#A 轉置矩陣
A.T
A.transpose()
```

```python
#A 反矩陣
np.linalg.inv(A)
```

# **NDarray** **統計功能**

```python
import numpy as np
normal_array = np.random.normal(0, 1, 10)
print(normal_array)
```

```python
## Min 
print(np.min(normal_array))

### Max 
print(np.max(normal_array))

### Mean 
print(np.mean(normal_array))

### 中數 Median
print(np.median(normal_array))

### 標準差 Sd
print(np.std(normal_array))
```
