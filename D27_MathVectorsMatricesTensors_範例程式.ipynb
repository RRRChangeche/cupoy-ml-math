
### 純量，向量，矩陣與張量 

### 1. 開始學 NumPy

確認程式庫程式庫版本

```python
import sys
import numpy as np

print ('Python: {}'.format(sys.version))
print ('NumPy: {}'.format(np.__version__))
```

認識純量(scaler)

```python
# a scalar
x = 6
x
```

在NumPy中,以陣列(Array)定義向量(Vector) 

```python
# a vector
x = np.array((1,2,3))
x
```

NumPy 矩陣可以定義形狀(shape)或大小(size).

```python
print ('Vector dimensions: {}'.format(x.shape))
print ('Vector size: {}'.format(x.size))
```

矩陣(matrix)的定義，運用  brackets 和 commas 分離各列. 

```python
x = np.matrix([[1,2,3],[4,5,6],[7,8,9]])
x
```

```python
print ('Matrix dimensions: {}'.format(x.shape))
print ('Matrix size: {}'.format(x.size))
```

利用NumPy定義3x3 矩陣 matrix:

```python
x = np.ones((3,3))
x
```

```python
print ('Matrix dimensions: {}'.format(x.shape))
print ('Matrix size: {}'.format(x.size))
```

雖然看不見，但 NumPy 可以很容易定義高維張量(tensor). 以下是如何建立3-D tensor. 4-D, 5-D, 和 6-D tensors 也一樣! 

```python
x = np.ones((3,3,3))
x
```

```python
print ('Tensor dimensions: {}'.format(x.shape))
print ('Tensor size: {}'.format(x.size))
```

### 2. 索引 (Indexing)

NumPy 的 索引 indexing 方式

```python
A = np.ones((5,5), dtype = np.int)
print (A)
```

```python
# indexing starts at 0
A[0,1] = 2
print (A)
```

```python
# NumPy用的是列(row) 行(column) 傳統
# 用 : 表示所有列或行
A[:,0] = 3
print (A)
```

```python
# 高維時增加索引
A = np.ones((5,5,5), dtype = np.int)

# 給第一列設值
A[:,0,0] = 6
print (A)
```

### 3. 矩陣運算 Matrix Operations

矩陣加法，減法和乘法

```python
A = np.matrix([[1,2],[3,4]])
B = np.ones((2,2), dtype = np.int)
```

```python
print (A)
```

```python
print (B)
```

```python
# 元素間加法
C = A + B
print (C)
```

```python
# 元素間減法
C = A - B
print (C)
```

```python
# 矩陣乘法
C = A*B
print (C)
```

### 4. 轉置矩陣 Matrix Transpose

轉置矩陣就是將元素行列位置互換.  

```python
A = np.array(range(9))
A = A.reshape(3,3)
print (A)
```

```python
# calculate tranpose of A
B = A.T
print (B)
```

```python
# taking the tranpose twice will return the original matrix
C = B.T
print (C)
```

```python
# The transpose will work arrays of any dimension
A = np.array(range(10))
A = A.reshape(2,5)
print (A)
```

```python
# create a column vector by taking the transpose
B = A.T
print (B)
```
