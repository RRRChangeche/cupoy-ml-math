# **NDarray** **統計功能**

```python
import numpy as np
normal_array = np.random.normal(5, 0.5, 10)
print(normal_array)
```

```python
## Min 
print(np.min(normal_array))

### Max 
print(np.max(normal_array))

### Mean 
print(np.mean(normal_array))

### Median
print(np.median(normal_array))

### Sd
print(np.std(normal_array))
```

```python
x = np.array([[4, 9, 2, 10],
        [6, 9, 7, 12]])
np.ptp(x, axis=1)
```

```python
np.ptp(x, axis=0)
```

```python
np.ptp(x)
```

# **NDArray 維度的運算**

3D Array

```python
import numpy as np
My3DArray = np.arange(12).reshape(2,2,3)
print (My3DArray)
```

3D Array contains 2D info

```python
import numpy as np
My3DArray1 = np.arange(4).reshape(2,2,1)
print (My3DArray1)
```

```python
My3DArray1 = np.array([[[0],[1]],[[2],[3]]])
print (My3DArray1)
```

2D Array

```python
import numpy as np
My2DArray = np.arange(4).reshape(2,2)
print (My2DArray)
```

```python
My2DArray = np.array([[0,1],[2,3]])
print (My2DArray)
```

Expand Dimension

```python
y = np.expand_dims(My2DArray, axis=0)
print (y)
```

```python
y = np.expand_dims(My2DArray, axis=1)
print (y)
```

```python
y = np.expand_dims(My2DArray, axis=2)
print (y)
```
