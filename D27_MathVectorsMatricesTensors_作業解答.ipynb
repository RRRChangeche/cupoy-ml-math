### I. 開始學 NumPy

確認程式庫程式庫版本

```
import sys
import numpy as np

print ('Python: {}'.format(sys.version))
print ('NumPy: {}'.format(np.__version__))
```

1.認識純量(scaler)
將純量60給 x

```
# a scalar
x = 60
x
```

2.在NumPy中,以陣列(Array)定義向量(Vector)
初始化向量[3, 4, 5] 

```
# a vector [3, 4, 5]
x = np.array((3,4,5))
x
```

3.NumPy 印出矩陣x形狀(shape)或大小(size)

```
print ('Vector dimensions: {}'.format(x.shape))
print ('Vector size: {}'.format(x.size))
```

4.矩陣(matrix)的定義，運用  brackets 和 commas 分離各列. 

```
x = np.matrix([[7,8,9],[4,5,6],[1,2,3]])
x
```

```
print ('Matrix dimensions: {}'.format(x.shape))
print ('Matrix size: {}'.format(x.size))
```

5.利用NumPy定義3x3 矩陣 matrix:

```
x = np.ones((4,4))
x
```

```
print ('Matrix dimensions: {}'.format(x.shape))
print ('Matrix size: {}'.format(x.size))
```

6.雖然看不見，但 NumPy 可以很容易定義高維張量(tensor). 以下是如何建立3-D tensor. 4-D, 5-D, 和 6-D tensors 也一樣! 

```
x = np.zeros((3,3,3))
x
```

```
print ('Tensor dimensions: {}'.format(x.shape))
print ('Tensor size: {}'.format(x.size))
```

### II. 索引 (Indexing)

1.NumPy 的 索引 indexing 方式

```
A = np.zeros((5,5), dtype = np.int)
print (A)
```

```
# indexing starts at 0
A[0,1] = 2
print (A)
```

```
# NumPy用的是列(row) 行(column) 傳統
# 用 : 表示所有列或行
A[:,0] = 3
print (A)
```

```
# 高維時增加索引
A = np.ones((5,5,5), dtype = np.int)

# 給第一列設值
A[:,0,0] = 6
print (A)
```

### III. 矩陣運算 Matrix Operations

1.矩陣加法，減法和乘法

```
A = np.matrix([[4,3],[2,1]])
B = np.ones((2,2), dtype = np.int)
```

```
print (A)
```

```
print (B)
```

```
# 元素間加法
C = A + B
print (C)
```

```
# 元素間減法
C = A - B
print (C)
```

```
# 矩陣乘法
C = A*B
print (C)
```

### IV. 轉置矩陣 Matrix Transpose

1.轉置矩陣就是將元素行列位置互換.  

```
A = np.array(range(16))
A = A.reshape(4,4)
print (A)
```

```
# 計算 A 的轉置
B = A.T
print (B)
```

```
# 轉置置兩次還原
C = B.T
print (C)
```

```
# 轉置可用在任何維度
A = np.array(range(10))
A = A.reshape(2,5)
print (A)
```

```
# 轉置產生行向量
B = A.T
print (B)
```
