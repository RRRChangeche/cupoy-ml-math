### 切分方程式

定義函數為 $$f(x) = x^2+1$$

```python
def f(x):
    return x**2+1
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5)
plt.plot(x,f(x))
plt.show()
```

f 微分後為$$f'(x)=2x$$

f 在 x=2 的切分方程式為 $$y=f'(2)(x-2)+f(2)\\=4x-3$$

```python
def f_c(x):
    return 4*x-3
```

```python
plt.plot(x,f(x),label='x^2+1')
plt.plot(x[25:],f_c(x[25:]),label='4x+3')
plt.legend()
plt.scatter(2,5,s=77)
plt.show()
```
