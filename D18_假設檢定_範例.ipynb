```python
import numpy as np
import matplotlib.pyplot as plt
```

#### 畫出 $\bar{X} \sim N(100,(10)^2/64)$

```python
def f(x,mu,sigma_squ):
    return 1/np.sqrt(2*sigma_squ*np.pi)*np.exp((-1/2)*(x-mu)**2/sigma_squ)
```

```python
x=np.linspace(95,105)
plt.plot(x,f(x,100,100/64))
plt.title(r'$\bar{X} \sim N\,(100,1.25)$')
plt.show()
```

```python
# 標準常態分布，查表會用到此概念
x = (x - np.mean(x))/np.std(x)
plt.plot(x,f(x,0,1))
plt.title(r'$N\,(0,1)$')
plt.show()
```
