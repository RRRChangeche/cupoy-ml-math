作業1：用Scipy 程式庫 計算餘弦相似度(Cosine Similarity)

```python
from scipy import spatial
List1 = [4, -47, 8, 3]
List2 = [3, 52, -12, 16]
result = 1 - spatial.distance.cosine(List1, List2)
print(result)
```

作業2：用Numpy 程式庫 計算餘弦相似度(Cosine Similarity)

```python
from numpy import dot
from numpy.linalg import norm
List1 = [4, 47, 8, -3]
List2 = [3, -52, 12, 16]
result = dot(List1, List2)/(norm(List1)*norm(List2))
print(result)
```

作業3：用 Scikit Learn 程式庫 計算餘弦相似度(Cosine Similarity)

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity,cosine_distances
A=np.array([10,4])
B=np.array([5,7])
result=cosine_similarity(A.reshape(1,-1),B.reshape(1,-1))
print(result)
```
